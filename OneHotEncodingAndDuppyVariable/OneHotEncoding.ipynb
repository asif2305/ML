```python
import pandas as pd
df=pd.read_csv('homeprices.csv')
df
```

# create dummies

```python
dummies=pd.get_dummies(df.town)
dummies
```

# concate dataframe

```python
merges=pd.concat([df,dummies],axis='columns')
merges
```

# Dummy Variable Trap
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

```python
final=merges.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

# create the training sets

```python
X=final.drop('price',axis='columns')
X
```

```python
y=final.price
y
```

# fit with the model

```python
model.fit(X,y)
```

# predict

```python
model.predict([[2800,0,0]])
```

```python
df['predict_price']=model.predict(X)
```

```python
df
```

# score

```python
model.score(X,y)
```

# Using sklearn OneHotEncoder
First step is to use label encoder to convert town names into numbers

# label encoder

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
dfle=df
```

```python
dfle
```

# fit transform

```python
dfle.town= le.fit_transform(dfle.town)
dfle
```

```python
dfle= dfle.drop('predict_price',axis=1)
dfle
```

# seperate x and y

```python
X= dfle[['town','area']].values
X
```

```python
y=dfle.price.values
y
```

# Now use one hot encoder to create dummy variables for each of the town

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct= ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
new_X=ct.fit_transform(X)
new_X
```

```python
new_X=new_X[:,1:]
new_X
```

```python
model.fit(new_X,y)
```

```python
model
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
